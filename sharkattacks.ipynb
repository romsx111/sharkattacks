```python
def categorize_activity(x):
    if pd.isna(x):
        return 'other'
    
    if 'surf' in x:
        return 'surfing'
    elif 'swim' in x or 'bathing' in x:
        return 'swimming'
    elif 'dive' in x or 'snorkel' in x:
        return 'diving/snorkeling'
    elif 'fish' in x:
        return 'fishing'
    elif 'boat' in x or 'kayak' in x or 'canoe' in x:
        return 'boating'
    elif 'wade' in x:
        return 'wading'
    elif 'stand' in x or 'paddle' in x:
        return 'paddleboarding'
    else:
        return 'other'
```

# DATE

```python
#Cleaned and formated the date column
import re
def format_gsaf_date(row):
    # Ensure we are working with strings and handle lowercase
    date_str = str(row['date']).lower()
    year = str(row['year'])
    
    # 1. Improved Day Pattern: Looks for 1 or 2 digits at the start of a word
    # \d{1,2} finds the numbers; (?:st|nd|rd|th)? ignores the suffix if present
    day_match = re.search(r'(\d{1,2})(?:st|nd|rd|th)?', date_str)
    
    # 2. Month Pattern: Standard 3-letter search
    month_match = re.search(r'(jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec)', date_str)
    
    if month_match and len(year) == 4:
        # Extract the day from the match
        day = day_match.group(1) if day_match else "01"
        month = month_match.group(1).capitalize()
        
        # Return as a string for pd.to_datetime to parse
        return f"{day}-{month}-{year}"
    
    return None
```

```python
# Apply the formatting
df['Clean_Date_Str'] = df.apply(format_gsaf_date, axis=1)
```

```python
# Convert to actual Pandas Datetime objects
df['date'] = pd.to_datetime(df['Clean_Date_Str'], format='%d-%b-%Y', errors='coerce')
```

```python
#Standardized case and removed extra spaces
df['type'] = df['type'].fillna('Unknown').str.strip().str.lower()
```

```python
#Mapped inconsistent values to standard categories
mapping = {
    'boat': 'boating',
    'watercraft': 'boating',
    'questionable': 'invalid',
    'unconfirmed': 'invalid',
    'unverified': 'invalid',
    'under investigation': 'invalid',
    '?': 'unknown'
}
def final_type(val):
    for key, value in mapping.items():
        if key == val:
            return value
    return val

df['type'] = df['type'].apply(final_type)
```

```python
# Aggregate data
country_stats = df.groupby('country').agg(total_attacks=('country','count'), fatalities=('injury', lambda x: (x == 'Fatal').sum())).reset_index()

country_stats['fatality_rate'] = country_stats['fatalities'] / country_stats['total_attacks']

# Focusing on top countries only
top_countries = country_stats.nlargest(10, 'total_attacks')

# Sorting for label placement
top_countries = top_countries.sort_values(by='total_attacks')

plt.figure(figsize=(13,8))

# Scatter
sns.scatterplot(data=top_countries,x='total_attacks',y='fatality_rate',size='total_attacks',sizes=(200, 800),alpha=0.7,edgecolor='black',linewidth=0.5,legend=False)

# Add labels (aligned better)
for i, row in top_countries.iterrows():
    plt.text(row['total_attacks'] + 2, row['fatality_rate'],row['country'],fontsize=10,va='center')

# Titles (clean hierarchy)
plt.title('Tourism Risk vs Exposure by Country', fontsize=16, weight='bold')
plt.suptitle('Where to Promote vs Where to avoid', fontsize=11, y=0.92)

plt.xlabel('Total Shark Attacks (Exposure)', fontsize=12)
plt.ylabel('Fatality Rate (Risk)', fontsize=12)

# Subtle grid + clean edges
plt.grid(alpha=0.2)
sns.despine()

plt.tight_layout()
plt.show()
```
