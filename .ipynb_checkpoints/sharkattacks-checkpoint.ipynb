```python
#Cleaning first steps
'''lowercase everything - Headers'''
df.columns =df.columns.str.lower()
df.tail(5)
```

```python
'''Drop completely empty lines '''

df = df.dropna(how ="all")
df.tail(5)
```

```python
'''Drop completly empty columns (Unnamed) '''

df = df.dropna(axis=1, how="all") #First double check if there's any othe with null and erase it

#Drop case number & case number.1
df= df.drop(columns=["unnamed: 21","unnamed: 22"])
df.tail(5)
```

```python
#Last 7 columns cleaning
'''' Source, pdf, href formula & href cleaning (Not needed for analysis)'''

df = df.drop(columns=["source","pdf","href formula", "href"])
df.tail(5)
```

```python
'''Case Number (review if its duplicated with Case Number.1 and leave 1)'''
#Review if they are the same:
(df["case number"]== df["case number.1"]).all()
```

```python
(df["case number"]== df["case number.1"]).sum()
```

```python
df[df["case number"]!= df["case number.1"]] #Here it takes also NaN values as different!
```

```python
#Review how many are different (Excluding NaNs)
df["case number"].equals(df["case number.1"])
```

```python
'''Use case number.1 as base to only leave one (when we have NaN values on case number.1 use case number to fill)'''
#Generate a final column of case number and name it case_number_final

df["case_number_final"] = df["case number.1"].fillna(df["case number"])
df.tail(7)
```

```python
#Erase case number and case number.1 and only leave the merged column of case_number_final
df = df.drop(columns=["case number", "case number.1"])
df.head(7)
```

```python
#Renamed case_number_final to the case_number
df=df.rename(columns={"case_number_final":"case_number"})
df.head(5)
```

```python
'''original order'''
#Removed the original order column - it does not provide analytical value + many null values
df = df.drop(columns=["original order"])
df.head(5)
```
